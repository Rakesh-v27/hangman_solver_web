```python
def filter_titles(titles, desired_word_count, not_in):
    """
    Filters a list of titles, returning only those that:
      - Have exactly `desired_word_count` words.
      - Do NOT contain any letter from the `not_in` list.
    
    Parameters:
        titles (list of str): The list of movie titles.
        desired_word_count (int): The exact number of words the title must have.
        not_in (list of str): Letters that must not appear in the title.
        
    Returns:
        list of str: The filtered list of titles.
    """
    filtered = []
    for title in titles:
        # Check if title has the desired word count
        if len(title.split()) != desired_word_count:
            continue
        
        # Convert title to lowercase for case-insensitive checking
        lower_title = title.lower()
        
        # Check if any letter in not_in is present in the title
        if any(letter in lower_title for letter in not_in):
            continue
        
        # If it passes both conditions, add it to the filtered list.
        filtered.append(title)
        
    return filtered

```

```python
import string

def get_letter_frequencies(movie_list):
    """
    Given a list of movie titles (all of the same word-count group),
    returns a dictionary mapping each letter (a-z) to the percentage of titles
    that contain that letter at least once.
    
    The dictionary is sorted in descending order of percentage.
    """
    total_movies = len(movie_list)
    if total_movies == 0:
        return {}
    
    # Normalize titles: convert each title to lowercase.
    normalized_titles = [title.lower() for title in movie_list]
    
    # Define the alphabet (only lowercase letters).
    alphabet = set(string.ascii_lowercase)
    
    # Count how many titles contain each letter.
    percentages = {}
    for letter in alphabet:
        count = sum(1 for t in normalized_titles if letter in t)
        percentages[letter] = (count / total_movies) * 100
    
    # Sort the dictionary by percentage in descending order.
    sorted_percentages = dict(sorted(percentages.items(), key=lambda x: x[1], reverse=True))
    
    return sorted_percentages
```

```python
def main_hangman_solver():
    """
    Main loop for a Hangman-style solver that uses three helper functions:
      1. movie_list(pattern: str) -> list
      2. filter_titles(candidates: list, word_count: list, not_in_list: list) -> list
      3. get_letter_frequencies(movie_list: list) -> dict
      
    The flow is:
      a) Prompt the user for the word lengths (e.g., "7,4" for a two-word movie with 7 and 4 letters).
      b) Build the initial pattern string (e.g., "_______.____").
      c) Get the candidate movie list using movie_list(pattern).
      d) Filter that candidate list using filter_titles() with the given word lengths and an initially empty not_in_list.
      e) Use get_letter_frequencies() on the filtered list to obtain a dictionary of letter frequencies.
      f) Suggest a letter (one not already in the current pattern) and ask the user if it is in the movie.
      g) If yes, ask for the positions (in a specified format) and update the pattern; if no, update not_in_list.
      h) Repeat until the pattern is complete.
    """
    
    # --- Step 1: Prompt for word structure ---
    word_lengths_input = input("Enter the letter counts for each word (comma-separated, e.g., '7,4'): ")
    # Parse input to a list of integers, e.g., [7, 4]
    try:
        word_lengths = [int(x.strip()) for x in word_lengths_input.split(',')]
    except ValueError:
        print("Invalid input. Please enter numbers separated by commas.")
        return
    
    print(word_lengths)

    # Build the initial pattern: for each word, create a string of underscores.
    current_pattern_words = ['_' * length for length in word_lengths]
    # Use a dot (.) as the separator between words.
    current_pattern = '.'.join(current_pattern_words)
    
    # not_in_list will hold letters that have been guessed as wrong.
    not_in_list = []
    
    solved = False
    
    while not solved:
        print("\nCurrent pattern:", current_pattern)
        
        # --- Step 2: Obtain candidate movies based on the current pattern ---
        # Here, we assume movie_list() accepts the current pattern and returns a candidate list.
        candidate_list = movie_list(current_pattern)
        
        # --- Step 3: Filter candidate list based on word lengths and wrong guesses ---
        filtered_candidates = filter_titles(candidate_list, word_lengths, not_in_list)
        
        # --- Step 4: Compute letter frequencies from the filtered candidate list ---
        freq_dict = get_letter_frequencies(filtered_candidates)
        
        # --- Step 5: Choose the next letter to guess ---
        # We want a letter that is not already in our current pattern.
        guessed_letter = None
        for letter, pct in freq_dict.items():
            if letter not in current_pattern:
                guessed_letter = letter
                break
        
        if guessed_letter is None:
            print("No new letters to guess. Stopping...")
            break
        
        print("Suggested letter to guess:", guessed_letter)
        user_response = input(f"Is the letter '{guessed_letter}' in the movie? (y/n): ").strip().lower()
        
        if user_response == 'y':
            # If the letter is present, ask for the positions.
            # Format: for each word, provide a comma-separated list of positions (1-indexed),
            # and separate words with a dash.
            # For example: "1,2-3" means in the first word the letter is at positions 1 and 2,
            # and in the second word the letter is at position 3.
            pos_input = input("Enter positions for the letter (use '-' to separate words, e.g., '1,2-3'): ").strip()
            
            # Split input by '-' to get positions for each word.
            pos_parts = pos_input.split('-')
            if len(pos_parts) != len(word_lengths):
                print("The number of position groups does not match the number of words.")
                continue  # Ask again in the next iteration.
            
            # Update current_pattern for each word.
            current_pattern_words = current_pattern.split('.')
            new_pattern_words = []
            for i, pos_str in enumerate(pos_parts):
                # Get the list of positions as integers (assume input is well-formed)
                try:
                    positions = [int(x.strip()) for x in pos_str.split(',') if x.strip().isdigit()]
                except ValueError:
                    print("Invalid position input. Try again.")
                    positions = []
                
                word_list = list(current_pattern_words[i])
                # Fill in the guessed letter at the indicated positions.
                for pos in positions:
                    if 1 <= pos <= len(word_list):
                        word_list[pos - 1] = guessed_letter
                new_pattern_words.append(''.join(word_list))
            
            # Update current_pattern with the newly filled letters.
            current_pattern = '.'.join(new_pattern_words)
        else:
            # If the letter is not in the movie, update the not_in_list.
            not_in_list.append(guessed_letter)
            print(f"Letter '{guessed_letter}' added to the not-in list.")
        
        # Check if the pattern is complete (i.e., no underscores remain)
        if '_' not in current_pattern:
            solved = True
            print("\nCongratulations! The movie is solved:")
            print(current_pattern)
            break
        
        # Optionally, print the number of remaining candidate movies for debugging.
        print(f"Remaining candidate movies: {len(filtered_candidates)}")
        
# -----------------------------------------------------
# The functions movie_list, filter_titles, and get_letter_frequencies
# are assumed to be defined elsewhere. For testing purposes, you might
# define dummy versions, for example:
#
# def movie_list(pattern):
#     # Dummy implementation; in practice, this would query your dataset.
#     return ["example movie", "another movie", ...]
#
# def filter_titles(candidate_list, word_lengths, not_in_list):
#     # Dummy filtering based on word lengths and not_in_list.
#     return [title for title in candidate_list if title_meets_conditions(title)]
#
# def get_letter_frequencies(movie_list):
#     # Dummy implementation returning a sample dictionary.
#     return {'e': 80.0, 't': 70.0, 's': 60.0, ...}
#
# -----------------------------------------------------
#
# Finally, call the main solver:
if __name__ == "__main__":
    main_hangman_solver()
```
